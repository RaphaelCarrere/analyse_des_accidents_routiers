```python
import pandas as pd

arbres = pd.read_csv("https://www.data.gouv.fr/api/1/datasets/r/60433484-f30e-44ef-a362-e5553a9b7a42", sep = ";")
```

```python
import os
import geopandas as gpd

path_iris = "data/iris.gpkg"

if os.path.exists(path_iris):
    # Lecture du fichier gpkg
    iris_france = gpd.read_file(path_iris)
    # Filtre sur Paris
    iris = iris_france[iris_france['code_insee'].str.startswith('75')].copy()
    print("Données de Paris chargées !")
else:
    print(f"ERREUR : Le fichier {path_iris} est absent.")
    print("Veuillez télécharger les données sur : https://cartes.gouv.fr/rechercher-une-donnee/dataset/IGNF_CONTOURS-IRIS?redirected_from=geoservices.ign.fr")
    print("Et placez le fichier .gpkg dans le dossier /data/")
```
